```python
delivery_df[['match_id','Ball_score']].groupby('match_id').cumsum()['Ball_score']
```

```python
delivery_df['Score'] = delivery_df[['match_id','Ball_score']].groupby('match_id').cumsum()['Ball_score']
```

```python
delivery_df.head()
```

```python
delivery_df[['Ball_score','Score']]
```

```python
# As the target is one run more than innings 1 score. 
delivery_df['target_left'] = (delivery_df['total_runs'] + 1) - delivery_df['Score'] 
delivery_df
```

```python
delivery_df['Remaining Balls'] = (120 - ((delivery_df['over'] - 1) * 6 + delivery_df['ball']))
delivery_df
```

```python
pd.set_option('display.max_columns', None)
```

```python
delivery_df.head()
```

```python
# Filling null values of player_dismissed with 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df
```

```python
# Replacing player value with 1 and converting it to int data type
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == '0' else '1').astype('int64')
```

```python
delivery_df[['match_id','player_dismissed']]
```

```python
# Number of Wickets fell.
delivery_df['Wickets'] = delivery_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df
```

```python
# Number of wickets left 
delivery_df['Wickets'] = 10 - delivery_df['Wickets']
delivery_df
```

#### Current Run Rate  

$$Current \ run \ rate = \frac{Runs \ Scored}{Number \ of \ Overs \ Completed}$$

```python
# crr = runs/overs --> Runs Scored by the number of Overs 
delivery_df['crr'] = (delivery_df['Score']*6)/(120 - delivery_df['Remaining Balls'])
```

```python
delivery_df
```

#### Required Run Rate  

$$Required \ run \ rate = \frac{Remaining \ Score \ or \ Target}{Number \ of \ Overs \ left}$$

```python
# rrr = runs/overs
delivery_df['rrr'] = (delivery_df['target_left']*6)/(delivery_df['Remaining Balls'])
```

```python
delivery_df
```

```python
# Function to say weither 2nd inning team won or not and 1 means they won
def result(row):
    if row['batting_team'] == row['winner']:
        return 1
    else: 
        return 0 
```

```python
# Applying this function on a a dataframe
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
# Splitting data to features and target.

X = Model_df.iloc[:,:-1] # all columns except result column and all rows

Y = Model_df.iloc[:,-1] # last column and all rows of it.
```

### Model Building

```python
# Importing Model required libraries
from sklearn.model_selection import train_test_split 

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2) 
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# Non Numerical columns are converted into numerical 
# One hot encoder we are converting it to vector representation as there is no hirerchy. 
# Transformer 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
],
remainder='passthrough')
```

```python
trf 
```

```python
# Building the logistic regression Model to predict 
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_prediction = pipe.predict(X_test)
```

# Evaluating the Model 

```python
# Accuracy score. 
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(Y_test, Y_prediction)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# Define models
models = {
    "Logistic Regression": LogisticRegression(solver='liblinear'),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1),
    "Decision Tree": DecisionTreeClassifier(max_depth=10, random_state=42),
}

results = {}

# Evaluate each model
for name, model in models.items():
    pipe = Pipeline([
        ('preprocessing', trf),  
        ('classifier', model)
    ])
    
    pipe.fit(X_train, Y_train)
    preds = pipe.predict(X_test)
    acc = accuracy_score(Y_test, preds)
    results[name] = acc
    print(f"{name}: {acc*100:.2f}%")
```

### ✅ Random Forest Classifier (Optimized)

- **Best Score**: 99.80% (via GridSearchCV)
- **Best Params**:  
  `n_estimators=200`, `max_depth=20`, `min_samples_split=2`
- **Trained on**: Google Colab (due to local hardware limits)
- **Model Ready** for deployment in Streamlit IPL Predictor 🚀


```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [5, 10, 20],
    'classifier__min_samples_split': [2, 5],
}

pipe_rf = Pipeline([
    ('preprocessing', trf),
    ('classifier', RandomForestClassifier(random_state=42))
])

grid = GridSearchCV(pipe_rf, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, Y_train)

print("Best Score:", grid.best_score_)
print("Best Params:", grid.best_params_)
```
